```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#1
img = cv2.imread('OldLenna.png')
plt.figure(figsize=(5,5))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
#2
axs = plt.subplots(nrows=3, ncols=3, figsize=(20,20))[1]

img_00 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
axs[0,0].set_title("grayscale")
axs[0,0].imshow(img_00, cmap=plt.cm.gist_gray)

img_01 = cv2.threshold(img_00, 128, 255, cv2.THRESH_BINARY)[1]
axs[0,1].set_title("THRESH_BINARY")
axs[0,1].imshow(img_01, cmap=plt.cm.gist_gray)

img_02 = cv2.threshold(img_00, 128, 255, cv2.THRESH_BINARY_INV)[1]
axs[0,2].set_title("THRESH_BINARY_INV")
axs[0,2].imshow(img_02, cmap=plt.cm.gist_gray)

img_10 = cv2.threshold(img_00, 128, 255, cv2.THRESH_TRUNC)[1]
axs[1,0].set_title("THRESH_TRUNC")
axs[1,0].imshow(img_10, cmap=plt.cm.gist_gray)

img_11 = cv2.threshold(img_00, 128, 255, cv2.THRESH_TOZERO)[1]
axs[1,1].set_title("THRESH_TOZERO")
axs[1,1].imshow(img_11, cmap=plt.cm.gist_gray)

img_12 = cv2.threshold(img_00, 128, 255, cv2.THRESH_TOZERO_INV)[1]
axs[1,2].set_title("THRESH_TOZERO_INV")
axs[1,2].imshow(img_12, cmap=plt.cm.gist_gray)

img_20 = cv2.threshold(img_00, 128, 255, cv2.THRESH_MASK)[1]
axs[2,0].set_title("THRESH_MASK")
axs[2,0].imshow(img_20, cmap=plt.cm.gist_gray)

img_21 = cv2.threshold(img_00, 128, 255, cv2.THRESH_OTSU)[1]
axs[2,1].set_title("THRESH_OTSU")
axs[2,1].imshow(img_21, cmap=plt.cm.gist_gray)

img_22 = cv2.threshold(img_00, 128, 255, cv2.THRESH_TRIANGLE)[1]
axs[2,2].set_title("THRESH_TRIANGLE")
axs[2,2].imshow(img_22, cmap=plt.cm.gist_gray)
```

```python
#3
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(img_hsv)
img_hsv_c = np.concatenate((h, s, v), axis=1)
plt.figure(figsize=(20, 20))
plt.imshow(img_hsv_c)
```
